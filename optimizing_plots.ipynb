```python
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
from matplotlib.colors import ListedColormap
```

```python
file = open('true_params', 'rb')
true_params=pkl.load(file)
file.close()

file = open('opt_params', 'rb')
opt_params=pkl.load(file)
file.close()

file = open('loss_history', 'rb')
loss_history=pkl.load(file)
file.close()

file=open('opt_data_ys', 'rb')
opt_solution_ys=pkl.load(file)
file.close()

file=open('opt_data_ts', 'rb')
opt_solution_ts=pkl.load(file)
file.close()

file=open('true_data_ys', 'rb')
true_solution_ys=pkl.load(file)
file.close()

file=open('true_data_ts', 'rb')
true_solution_ts=pkl.load(file)
file.close()
```

```python
species=['W', 'WE1', 'W*', 'W*E2', 'E1', 'E2']
```

```python
cmap = plt.colormaps['Pastel1']
colors = cmap(np.linspace(0, 1, 9))
for i in range(true_solution_ys.shape[1]):
    plt.plot(true_solution_ts, true_solution_ys[:, i], color=colors[i], label=f'{species[i]}, true params')
    plt.plot(opt_solution_ts, opt_solution_ys[:, i], color=colors[i], linestyle='--', label='opt params')

plt.title(f'true params:a1={true_params[0]:0.2f}, a2={true_params[1]:0.2f}, d1={true_params[2]:0.2f}, d2={true_params[3]:0.2f}, k1={true_params[4]:0.2f}, k2={true_params[5]:0.2f} \n opt params:a1={opt_params[0]:0.2f}, a2={opt_params[1]:0.2f}, d1={opt_params[2]:0.2f}, d2={opt_params[3]:0.2f}, k1={opt_params[4]:0.2f}, k2={opt_params[5]:0.2f}')
plt.legend(bbox_to_anchor=[1.05, 1], loc='upper left')
```
